## Задание 1
Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.


```python
import requests
import json
from pprint import pprint
```

```python
# Так как можно отправлять 60 запросов без регистрации, на прямую пишем запрос
url = 'https://api.github.com/users/xenia-canary/repos'

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.135 Safari/537.36'}
res = requests.get(url, headers = headers)
j_data = res.json()

with open('to_json.json', 'w') as f:
    f.write(json.dumps(j_data))

with open('to_json.json') as f:
    pprint(f.read())
```
